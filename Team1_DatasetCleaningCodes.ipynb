```python
import os
import numpy as np
import pandas as pd

#os.chdir("C:\\Users\\jeffr\\Downloads")
df = pd.read_csv("Life Expectancy Data.csv")
df.columns
```

```python
#Joining Population and GDP data

pop=pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_63973.csv")
gdp=pd.read_csv("API_NY.GDP.MKTP.CD_DS2_en_csv_v2_40924.csv")

df.drop('Population',axis=1,inplace=True)
df.drop('GDP',axis=1,inplace=True)

pop1=pd.melt(pop,id_vars=['Country Name'],var_name='Year')

convert_dict = {'Year':int}
pop1=pop1.astype(convert_dict)
#print(pop1.dtypes)

pop1.rename(columns={'Country Name':'Country'},inplace=True)

df2=pd.merge(df,pop1,on=['Country','Year'],how='left')
df2.rename(columns={'value':'Population'},inplace=True)

gdp1=pd.melt(gdp,id_vars=['Country Name'],var_name='Year')

convert_dict = {'Year':int}
gdp1=gdp1.astype(convert_dict)

gdp1.rename(columns={'Country Name':'Country'},inplace=True)

df_clean=pd.merge(df2,gdp1,on=['Country','Year'],how='left')
df_clean.rename(columns={'value':'GDP'},inplace=True)
df_clean_map=df_clean

df_interpolated = pd.DataFrame()
```

```python
#Removing countries which have null values (all 16 years) for many columns  

df_clean = df_clean[~df_clean.Country.isin (['Republic of Moldova','Bolivia (Plurinational State of)','Gambia','South Sudan','Palau','Swaziland','Niue','Democratic People''s Republic of Korea','Cook Islands','CÃ´te d''Ivoire','Bolivia (Plurinational State of)'])]
df_clean = df_clean[np.isfinite(df_clean['GDP'])]

#Filling missing values for alcohol consumption based on the mean aggregated at country level. For instance, if a country does not have alcohol consumption for any year, the blanks will be replaced by the aggregated mean at coutnry level  

df_clean['Alcohol'] = df_clean['Alcohol'].fillna(df_clean.groupby('Country')['Alcohol'].transform('mean'))
 
df_clean['Hepatitis B'] = df_clean['Hepatitis B'].fillna(df_clean.groupby('Status')['Hepatitis B'].transform('mean'))
df_clean[' thinness  1-19 years'] = df_clean[' thinness  1-19 years'].fillna(df_clean.groupby('Status')[' thinness  1-19 years'].transform('mean'))
df_clean[' thinness 5-9 years'] = df_clean[' thinness 5-9 years'].fillna(df_clean.groupby('Status')[' thinness 5-9 years'].transform('mean'))
df_clean['Adult Mortality'] = df_clean['Adult Mortality'].fillna(df_clean.groupby('Status')['Adult Mortality'].transform('mean'))
df_clean['Income composition of resources'] = df_clean['Income composition of resources'].fillna(df_clean.groupby('Status')['Income composition of resources'].transform('mean'))
df_clean['Schooling'] = df_clean['Schooling'].fillna(df_clean.groupby('Status')['Schooling'].transform('mean'))
df_clean[' BMI '] = df_clean[' BMI '].fillna(df_clean.groupby('Status')[' BMI '].transform('mean'))
df_clean['Total expenditure'] = df_clean['Total expenditure'].fillna(df_clean.groupby('Status')['Total expenditure'].transform('mean'))

df_clean_map.to_csv("Charts_Data.csv")

list_country = list(df_clean['Country'].unique())

df_clean.loc[df_clean['Country']=="France"]
```

```python
#Removing Outliers for Column Hepatitis B and setting them as blanks

#Calculating mean and standard deviation of Hep B aggregated @Country level & merging 2 data frames into one
df4=df_clean.groupby('Country')['Total expenditure','Adult Mortality','Hepatitis B','Measles ','Schooling','Polio','Diphtheria '].mean()
df5=df_clean.groupby('Country')['Total expenditure','Adult Mortality','Hepatitis B','Measles ','Schooling','Polio','Diphtheria '].std()

df6=pd.merge(df4,df5, on=['Country'])

#Calculating the upper and lower limits for Hep B
df6['mean_plus_3sigma']=df6['Total expenditure_x']+df6['Total expenditure_y']*3
df6['mean_minus_3sigma']=df6['Total expenditure_x']-df6['Total expenditure_y']*1

df6['mean_plus_3sigma_AM']=df6['Adult Mortality_x']+df6['Adult Mortality_y']*3
df6['mean_minus_3sigma_AM']=df6['Adult Mortality_x']-df6['Adult Mortality_y']*1

df6['mean_plus_3sigma_HepB']=df6['Hepatitis B_x']+df6['Hepatitis B_y']*3
df6['mean_minus_3sigma_HepB']=df6['Hepatitis B_x']-df6['Hepatitis B_y']*1

df6['mean_plus_3sigma_Measles']=df6['Measles _x']+df6['Measles _y']*3
df6['mean_minus_3sigma_Measles']=df6['Measles _x']-df6['Measles _y']*1

df6['mean_plus_3sigma_Schooling']=df6['Schooling_x']+df6['Schooling_y']*3
df6['mean_minus_3sigma_Schooling']=df6['Schooling_x']-df6['Schooling_y']*1

df6['mean_plus_3sigma_Polio']=df6['Polio_x']+df6['Polio_y']*3
df6['mean_minus_3sigma_Polio']=df6['Polio_x']-df6['Polio_y']*1

df6['mean_plus_3sigma_Diphtheria']=df6['Diphtheria _x']+df6['Diphtheria _y']*3
df6['mean_minus_3sigma_Diphtheria']=df6['Diphtheria _x']-df6['Diphtheria _y']*1

#df6['mean_plus_3sigma_HIV']=df6[' HIV/AIDS_x']+df6[' HIV/AIDS_y']*3
#df6['mean_minus_3sigma_HIV']=df6[' HIV/AIDS_x']-df6[' HIV/AIDS_y']*1

#Merging(Joining) the calculated upper and lower limits with the original table
df7=pd.merge(df_clean,df6,on=['Country'])

#Comparing existing numbers with the upper and lower ranges and creating a flag for outliers
df7['is_outlier'] = np.where((df7['Total expenditure']>=df7['mean_plus_3sigma'])|(df7['Total expenditure']<=df7['mean_minus_3sigma']), 'yes', 'no')
df7['is_outlier_AM'] = np.where((df7['Adult Mortality']>=df7['mean_plus_3sigma_AM'])|(df7['Adult Mortality']<=df7['mean_minus_3sigma_AM']), 'yes', 'no')
df7['is_outlier_HepB'] = np.where((df7['Hepatitis B']>=df7['mean_plus_3sigma_HepB'])|(df7['Hepatitis B']<=df7['mean_minus_3sigma_HepB']), 'yes', 'no')
df7['is_outlier_Measles'] = np.where((df7['Measles ']>=df7['mean_plus_3sigma_Measles'])|(df7['Measles ']<=df7['mean_minus_3sigma_Measles']), 'yes', 'no')
df7['is_outlier_Schooling'] = np.where((df7['Schooling']>=df7['mean_plus_3sigma_Schooling'])|(df7['Schooling']<=df7['mean_minus_3sigma_Schooling']), 'yes', 'no')
df7['is_outlier_Polio'] = np.where((df7['Polio']>=df7['mean_plus_3sigma_Polio'])|(df7['Polio']<=df7['mean_minus_3sigma_Polio']), 'yes', 'no')
df7['is_outlier_Diphtheria'] = np.where((df7['Diphtheria ']>=df7['mean_plus_3sigma_Diphtheria'])|(df7['Diphtheria ']<=df7['mean_minus_3sigma_Diphtheria']), 'yes', 'no')

df7['Total expenditure_Updated'] = np.where((df7['is_outlier']=='yes'), " " , df7['Total expenditure'])
df7['Adult Mortality_Updated'] = np.where((df7['is_outlier_AM']=='yes'), " " , df7['Adult Mortality'])
df7['Hepatitis B_Updated'] = np.where((df7['is_outlier_HepB']=='yes'), " " , df7['Hepatitis B'])
df7['Measles_Updated'] = np.where((df7['is_outlier_Measles']=='yes'), " " , df7['Measles '])
df7['Schooling_Updated'] = np.where((df7['is_outlier_Schooling']=='yes'), " " , df7['Schooling'])
df7['Polio_Updated'] = np.where((df7['is_outlier_Polio']=='yes'), " " , df7['Polio'])
df7['Diphtheria_Updated'] = np.where((df7['is_outlier_Diphtheria']=='yes'), " " , df7['Diphtheria ']) 

df7['Total expenditure_Updated'] = df7['Total expenditure_Updated'].replace({" ":np.nan, " ":np.nan})
df7['Adult Mortality_Updated'] = df7['Adult Mortality_Updated'].replace({" ":np.nan, " ":np.nan})
df7['Hepatitis B_Updated'] = df7['Hepatitis B_Updated'].replace({" ":np.nan, " ":np.nan})
df7['Measles_Updated'] = df7['Measles_Updated'].replace({" ":np.nan, " ":np.nan})
df7['Schooling_Updated'] = df7['Schooling_Updated'].replace({" ":np.nan, " ":np.nan})
df7['Polio_Updated'] = df7['Polio_Updated'].replace({" ":np.nan, " ":np.nan})
df7['Diphtheria_Updated'] = df7['Diphtheria_Updated'].replace({" ":np.nan, " ":np.nan})
```
